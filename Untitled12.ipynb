```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
# import data from csv file. upload this file in google colab files in left menu under sample_data
customer_data = pd.read_csv('/content/Mall_Customers.csv')
```

```python
x=customer_data.iloc[:,[3,4]].values
```

```python
#dendrogram to find optimal no of clusters

import scipy.cluster.hierarchy as hie
dendrogram=hie.dendrogram(hie.linkage(x,method='ward'))
plt.xlabel('Customers')
plt.ylabel('Euc Dis')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ac=AgglomerativeClustering(n_clusters=5, affinity="euclidean", linkage="ward")
res=ac.fit_predict(x)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x[res==0,0], x[res==0,1], c='green', label='Cluster 1-High Income-low Spending')
plt.scatter(x[res==1,0], x[res==1,1], c='red', label='Cluster 2-low Income-low Spending')
plt.scatter(x[res==2,0], x[res==2,1], c='black', label='Cluster 3-High Income-low Spending')
plt.scatter(x[res==3,0], x[res==3,1], c='violet', label='Cluster 4-High Income-low Spending')
plt.scatter(x[res==4,0], x[res==4,1], c='blue', label='Cluster 5-low Income-low Spending')

plt.title('customer group')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
