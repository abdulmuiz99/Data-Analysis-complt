```python
import pandas as pd
import seaborn as sns
sns.set_theme(style="ticks", color_codes="True")
```

```python
df = pd.read_csv("drugs_for_common_treatments.csv")
df
```

```python
sns.histplot (data=df, x="drug_name", hue="pregnancy_category", multiple="stack")
```

```python
sns.displot(data=df, x="medical_condition", hue="pregnancy_category", col="pregnancy_category")
```

```python
sns.pairplot(df, hue="pregnancy_category")
```

```python
import matplotlib
from matplotlib import pyplot as plt
```

```python
plt.bar(df['medical_condition'], df['activity'])
plt.show()
```

```python
sns.boxplot(x="rating", hue="pregnancy_category", data=df)
sns.despine(offset=10, trim=True)
```

```python
g = sns.PairGrid(df)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(df, hue="pregnancy_category")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
```
