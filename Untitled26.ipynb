```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
myretaildata = pd.read_excel('https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx')
```

```python
myretaildata.head(10)
```

```python
import matplotlib.pyplot as plt
myretaildata[['InvoiceDate', 'Description']].groupby('InvoiceDate').count()\
.plot(figsize=(12,6), fontsize=18)
plt.title('Number of Items Sold per Day', fontsize=18)
plt.xlabel('date', fontsize=14)
plt.ylabel('Qty', fontsize=14)
```

```python
myretaildata['Description'] = myretaildata['Description'].str.strip()
myretaildata.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
myretaildata['InvoiceNo'] = myretaildata['InvoiceNo'].astype('str')
myretaildata = myretaildata[~myretaildata['InvoiceNo'].str.contains('c')]
myretaildata.head()
```

```python
myretaildata['Country'].value_counts()
```

```python
mybasket = (myretaildata[myretaildata['Country'] == "Spain"]
            .groupby(['InvoiceNo', 'Description'])['Quantity']
            .sum().unstack().reset_index().fillna(0)
            .set_index('InvoiceNo'))
```

```python
def my_encode_units(x):
  if x <= 0:
    return 0
  if x >= 1:
    return 1

my_basket_sets = mybasket.applymap(my_encode_units)
my_basket_sets.drop('POSTAGE', inplace=True, axis=1)
```

```python
my_frequent_itemsets = apriori(my_basket_sets, min_support=0.05, use_colnames=True)
my_rules = association_rules(my_frequent_itemsets, metric="confidence", min_threshold=0.8)
```
