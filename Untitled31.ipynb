```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url=("https://www.pakwheels.com/used-bikes/search/-/")
p=requests.get(url)
htmlcontent=p.content
soup=BeautifulSoup(htmlcontent,'html.parser')
soup
```

```python
soup.prettify
```

```python
df=pd
row1=soup.find_all('strong',class_='generic-white fs18')
price = []
for p in row1:
    price.append(p.text)
price
```

```python
row2=soup.find_all('div',class_='search-title')
title = []
for n in row2:
    title.append(n.text.replace('\n',"")[1:-1])
title
```

```python
row3=soup.find_all('div',class_='clearfix')
Sellers_Comments = []
for n in row3:
    Sellers_Comments.append(n.text.replace('\n',""))
Sellers_Comments 
```

```python
df=pd.DataFrame(row1,columns=['price'])
df['title']=row2
df.to_csv('pakwheels.csv')
```
